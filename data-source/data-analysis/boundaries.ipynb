```python
import geopandas as gpd 
import fiona 
import numpy as np
import pandas as pd 
```

```python
gdf_wards = gpd.read_parquet('data/Jaipur_Wards.geoparquet')
gdf_muni_bounds = gpd.read_parquet('data/Jaipur_Boundary.geoparquet')
```

```python
gdf_wards.to_crs('EPSG:4326', inplace=True)
gdf_muni_bounds.to_crs('EPSG:4326', inplace=True)
```

```python
gdf_muni_bounds.plot()
```

```python
gdf_wards.to_file('../../public/data/mapMain/jaipur_wards.geojson', driver='GeoJSON')
gdf_muni_bounds.to_file('../../public/data/mapMain/jaipur_muni_bounds.geojson', driver='GeoJSON')
```
