```python
import pandas as pd
df = pd.read_csv('data/SPA_DataCatalogue_condensed(in).csv')
```

```python
df.head()
```

```python
# Filter columns and rename
df = df[[
    'id',
    'data_display_name',
    'description',
    'theme',
    'geog_level2',
    'format2',
    'provider',
    'reference',
    'reference table/fig/pgno',
    'latest_year',
    'update_freq2',
    'url'
]]

df =df.rename(columns={
    'data_display_name': 'name',
    'geog_level2': 'resolution',
    'format2': 'format',
    'reference table/fig/pgno': 'reference_detail',
    'update_freq2': 'update_frequency'
})

df.head()
```

```python
# check for missing values
df.isna().sum()
```

```python
# Clean up missing value
df['update_frequency'] = df['update_frequency'].replace('unknown', 'N/A')# Update frequency unknown to N/A
df['reference_detail'] = df['reference_detail'].fillna('N/A')  # Fill missing reference details with 'N/A'
df['latest_year'] = df['latest_year'].fillna(0).astype(int) # Fill missing latest year with 0
df['latest_year'] = df['latest_year'].replace(0, 'N/A')  # Replace 0 with 'N/A' for latest year
df.head()
```

```python
# Change case for certain columns
df['name'] = df['name'].str.upper()
df['theme'] = df['theme'].str.title()
df['resolution'] = df['resolution'].str.capitalize()
df['format'] = df['format'].str.capitalize()
df.head()
```

```python
# Add a new row for Jaipur Census Data 2011 (new sample data) 
df = pd.concat([df, pd.DataFrame([
                  {
                      'id': df['id'].max() + 1,
                      'name': 'JAIPUR CENSUS DATA 2011',
                      'description': 'Census data for Jaipur city from the 2011 Census of India, including demographic details such as population, literacy rates, and household information.',
                      'theme': 'Demographic Profile',
                      'resolution': 'Ward',
                      'format': 'ZIP',
                      'provider': 'National Statistics Office',
                      'reference': 'Government of India',
                      'latest_year': 2011,
                      'update_frequency': 'Every 10 years',
                      'url_sample': 'https://orca.casa.ucl.ac.uk/~jens/sparc-dash/download/census2011.zip',
                      'url': 'https://censusindia.gov.in/'
                  }
              ])], ignore_index=True)
df.tail()
```

```python
# Add another column for latest_year_grouped in to last 2 years, 5 years, 10 years, and more than 10 years
import numpy as np
import datetime
current_year = datetime.datetime.now().year

def lastupdated_group(year, yearlist=[2,5,10,15,20]):
    if year == 'N/A':
        return 'N/A'
    year = int(year)
    counter = 0
    for i in yearlist:
        if current_year - year <= i:
            return f'Last {i} yrs'
        counter += 1
    return f'{yearlist[counter-1]+1} yrs or older'
  
df['lastupdate'] = df['latest_year'].apply(lastupdated_group)  
                                              
df.head()
```

```python
df.to_json('../../public/data/datacatalogue.json', orient='records', indent=4)
```
