```python
import pandas as pd
df = pd.read_csv('datacatalogue.csv',encoding='utf-8')
df.head()
```

```python
df = df[[
    'id',
    'data_display_name',
    'description',
    'format',
    'provider',
    'latest_year',
    'geog_level2',
    'theme',
    'url'
]]
```

```python
df = df.rename(columns={
    'data_display_name': 'name',
    'geog_level2': 'coverage',
    'url': 'fileUrl'
    })
df.head()
```

```python
# COnvert to json records
df.to_json('datacatalogue.json',  orient='records', indent=2)
```

```python
df.to_csv('datacatalogue-cleaned.csv', index=False, encoding='utf-8')
```
